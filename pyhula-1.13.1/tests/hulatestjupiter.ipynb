```python
# Step 1: Uninstall PyHula completely
import sys
import subprocess

subprocess.run([sys.executable, "-m", "pip", "uninstall", "pyhula", "-y"])
if 'pyhula' in sys.modules:
    del sys.modules['pyhula']
```

```python
# Step 2: Reinstall PyHula from Python 3.13 compatible wheel
import subprocess
import sys
import os

print(" Installing PyHula Python 3.13 compatible version...")
print("=" * 50)

# Path to our custom built wheel
wheel_path = r"C:\Users\janis\OneDrive\Dokumente\01-projects\Hula-python\pyhula-win11\pyhula-1.13.1\dist\pyhula-1.13.1-cp313-cp313-win_amd64.whl"

if os.path.exists(wheel_path):
    print(f"[OK] Found wheel: {wheel_path}")
    
    try:
        # Install the wheel
        result = subprocess.run([
            sys.executable, "-m", "pip", "install", wheel_path, "--force-reinstall"
        ], capture_output=True, text=True)
        
        print("INSTALL OUTPUT:")
        print(result.stdout)
        if result.stderr:
            print("INSTALL WARNINGS:")
            print(result.stderr)
        
        if result.returncode == 0:
            print("[OK] PyHula installed successfully!")
            
            # Test import
            try:
                import pyhula
                print("[OK] PyHula import test successful!")
                print(f"PyHula location: {pyhula.__file__}")
                
                # Test core modules
                from pyhula.pypack.fylo import commandprocessor
                from pyhula.pypack.system import datacenter
                print("[OK] Core modules imported successfully!")
                
            except ImportError as e:
                print(f"[WARNING] Import test failed: {e}")
                
        else:
            print("[ERROR] Installation failed")
            
    except Exception as e:
        print(f"[ERROR] Error during installation: {e}")
        
else:
    print(f"[ERROR] Wheel file not found: {wheel_path}")
    print("Please make sure the wheel was built successfully in the previous steps.")
```
